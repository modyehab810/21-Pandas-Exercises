### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
def group_by_decade(df):
    new_index = []
    decade_df = pd.DataFrame(columns=df.columns)

    for i in range(df.shape[0]):
        if i % 10 == 0:
            new_index.append(df.index[i])

    for i in range(len(new_index)):
        temp = np.array(df.loc[new_index[i]: new_index[i]+np.timedelta64(10,"Y")].sum())
        decade_df.loc[i]  = temp
        # Line To Get The Max Population From Decade Years
        decade_df.loc[i, "Population"] = np.array(df.loc[new_index[i]: new_index[i]+np.timedelta64(10,"Y"), "Population"].max())

    decade_df["Decade"] = new_index
    decade_df.set_index("Decade", inplace=True)
    return decade_df
```

```python
new_decade_df = group_by_decade(crime)
new_decade_df
```
